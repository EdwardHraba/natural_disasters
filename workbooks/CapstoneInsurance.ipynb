```python
insurance_avg = pd.read_csv('../data/state_averages.csv')
```

```python
insurance_avg_by_home_price = pd.read_csv('../data/state_avg_by_home_price.csv')
```

```python
population = pd.read_csv('../data/census.csv')
```

```python
insurance_averages = pd.merge(insurance_avg, insurance_avg_by_home_price,
         left_on = 'State',
         right_on = 'State',
         how = 'outer')
```

```python
insurance_averages['Difference From National Avg'] = insurance_averages['Difference From National Avg'].str.replace('(','').str.replace(')','')
```

```python
my_df = pd.merge(insurance_averages,population,
         left_on = 'State',
         right_on = 'State',
         how = 'outer')
```
